```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import numpy as np 
```

```python
x = np.linspace(0,10,1000)
```

```python
plt.rcParams.update({'font.size' : 15})
plt.plot(x,np.sin(x), color='black',linestyle='dotted', label='sin(x)')
plt.axis([0,10,-1,1])
plt.title('First graph')
plt.legend(loc='lower left')
ax = plt.axes()
ax.set(xlabel='x',ylabel='sin(x)')
```

```python
x1 = np.linspace(0,10,50)
```

```python
dy = 0.2
y1 = np.sin(x1) + dy*np.random.randn(50)
```

```python
plt.errorbar(x1,y1,yerr=dy,fmt='.k')
```

```python
%matplotlib inline 
import matplotlib.pyplot as plt 
import numpy as np 

fig = plt.figure(figsize=(12,8))

x = np.linspace(0,10,1000)
y = np.sin(x)

for i in range(6) : 
    fig.add_subplot(3,2,i+1) # graphe numeroté a partir de 1 
    plt.style.use(plt.style.available[i])
    plt.plot(x,y)
#    plt.text(s=plt.style.available[i],x=5,y=2,color='red')
    plt.text(s=plt.style.available[i],x=5.,y=1.2,color='red')


plt.show()
```
